```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:/Users/phari/Desktop/project_ml/Dataset.csv")
```

```python
df["Cuisines"].fillna("Unknown", inplace=True)
```

```python
df = df[[
    "Restaurant Name",
    "Cuisines",
    "Average Cost for two",
    "Price range",
    "Aggregate rating"
]]
```

```python
df.dropna(inplace=True)
```

```python
le = LabelEncoder()
df["Cuisine_Encoded"] = le.fit_transform(df["Cuisines"])
```

```python
preferred_cuisine = "North Indian"
preferred_price_range = 2
```

```python
recommendations = df[
    (df["Cuisines"].str.contains(preferred_cuisine, case=False, na=False))
    &
    (df["Price range"] == preferred_price_range)
]
```

```python
recommendations = recommendations.sort_values(
    by="Aggregate rating",
    ascending=False
)
```

```python
top_recommendations = recommendations.head(10)
```

```python
print("\nRecommended Restaurants\n")
```
